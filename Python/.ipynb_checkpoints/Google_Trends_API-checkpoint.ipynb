# Test Google Trends API

```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ['recession', 'debt', 'color', 'stocks', 'restaurant', 'portfolio', 'inflation', 'housing', 'dow jones', 'revenue']
kw_list = ['economics', 'credit', 'markets', 'return']
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='US', gprop='news') #today 5-y
```

```python
hits_recession = pytrends.interest_over_time()
```

```python
hits_recession.to_csv('news_google_9.csv')
```
