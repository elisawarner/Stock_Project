# Create Feature Set

```python
import pandas as pd
from datetime import datetime, timedelta
```

### Convert Date

```python
def convert_date(df, datecolname):
    converted_list = []

    original_list = df[datecolname].values
    
    for i in range(len(original_list)):
        try:
            date = datetime.strptime(original_list[i], '%Y-%m-%d')
        except:
            date = datetime.strptime(original_list[i], '%Y-%m')
        
        converted_list.append(date.strftime('%-m/%-d/%y'))

    df[datecolname] = converted_list
    
    return df
```

### Fill Data

```python
def fill_data(df, datecolname):
    new_df = pd.DataFrame()

    original_list = df[datecolname].values
    #print(original_list)
    
    date = datetime.strptime(original_list[0], '%m/%d/%y')

    j = 1
    current_date = date
    #print(current_date)
    #print(date.year)
    for i in range(365 * (2020 - date.year)):
        #print(i)
        df_date = datetime.strptime(original_list[j], '%m/%d/%y')

        #print(current_date, df_date)
        
        if current_date == df_date:
            new_df = new_df.append(df.iloc[j], ignore_index = True)
            
            if j < len(original_list) - 1:
                j += 1
            else:
                j = len(original_list) - 1
                print(original_list[j])
        elif current_date < df_date:
            row = df.iloc[j - 1]
            row = row.replace(row[datecolname], current_date.strftime('%-m/%-d/%y')) # adjust the date
            
            new_df = new_df.append(row, ignore_index = True) # append new row with new date to new dataframe

        current_date += timedelta(days=1)
        
    #    print(date)

    return new_df
```

### Remove Missing

```python
def remove_missing(df):
    for col in list(df):
        colvals = df[col].values
        new_colvals = []
        for i in range(len(colvals)):
            if colvals[i] == '.':
                new_colvals.append(colvals[i-1])
            else:
                new_colvals.append(colvals[i])
        df[col] = new_colvals
    
    return df
```
